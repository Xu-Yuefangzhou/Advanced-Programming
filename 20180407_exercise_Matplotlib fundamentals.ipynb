## 啊啊啊！其实不用导入random！
- rand是numpy里面的
- 这里建议不要按照这种方式导入：from numpy import ；应该按这样导入：import numpy as np，然后用np.random.rand调用，因为模块之间会冲突！

## imshow 显示图片

````python
from scipy.misc import ascent
img=ascent()
img
````

````python
imshow(img,extent=[-25,25,-25,25])
colorbar()
````

````python
imshow(img,extent=[-25,25,-25,25],cmap=cm.bone)
colorbar()
````

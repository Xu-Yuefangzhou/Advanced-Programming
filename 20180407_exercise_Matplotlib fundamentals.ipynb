## imshow 

````python
from scipy.misc import ascent
img=ascent()
img
````

````python
imshow(img,extent=[-25,25,-25,25])
colorbar()
````

````python
imshow(img,extent=[-25,25,-25,25],cmap=cm.bone)
colorbar()
````

````python
from scipy.misc import face
img=face()
img
imshow(img,extent=[-25,25,-25,25],cmap=cm.bone)
colorbar()
````

```python
%matplotlib inline
x=np.random.rand(200)
color=[1,1,1]
scatter(x,sin(x),50,color)
```

```python
%matplotlib inline
x=np.random.rand(200)
color=[1,1,0.5]
color2=[1,0,0.5]
color3=[1,0.5,0.5]
color4=[1,0.5,1]
color5=[1,0,1]
scatter(x,sin(x),70,color)
scatter(x,sin(x)+0.1,70,color2)
scatter(x,sin(x)+0.2,70,color3)
scatter(x,sin(x)+0.3,70,color4)
scatter(x,sin(x)+0.4,70,color5)
```
